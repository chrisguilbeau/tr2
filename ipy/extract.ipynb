```python
import bs4
from bs4 import BeautifulSoup
from urllib import urlopen
```

```python
page = urlopen('http://www.apostolic-churches.net/bible/strongs/jeremiah_29.shtml')
soup = BeautifulSoup(page.read(), 'html.parser')
```

```python
def hrefToRef(href):
    def getPre():
        if 'hebrew' in href:
            return 'H'
        else:
            return 'G'
    def getNum():
        return href.split('=')[-1].lstrip('0')
    return '{}{}'.format(getPre(), getNum())

def getRows(bookCode):
    for tr in soup.find_all('tr')[:5]:
        if tr.find('dt'):
            dt = tr.find('dt')
            b = dt.b
            for i, t in enumerate(b.next_siblings):
                def getCols():
                    if t and t.string and t.string.strip():
                        cparts = b.string.strip().split(':')
                        yield bookCode
                        yield cparts[0]
                        yield cparts[1]
                        yield t.string.strip()
                        if isinstance(t, bs4.element.Tag) and t.name == 'a':
                            href = t.get('href')
                            yield hrefToRef(href)
                        else:
                            yield None
                yield list(getCols())
                
for row in getRows('genesis'):
    if row:
        print row
```

```python
def getRows():
    for i, (test, book, chap_count) in enumerate(getBooks()):
        yield book, book.replace('_', ' ').title().replace(
            'Of','of'), i + 1, test, chap_count
books = pd.DataFrame(getRows())
books.columns = ('book', 'name', 'order', 'testament', 'chap_count')
books.to_csv('csv/book.csv', index=False)
books
```
